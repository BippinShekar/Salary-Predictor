## Data Cleaning

```python
#Converting all string enntries into Digit entries for passing through algorithhms
def find_and_convert(data):
    for i in data:
        x_to_find = data[i].values
        if np.char.isdigit(str(x_to_find[0])) is True:
            continue
        elif np.issubdtype(x_to_find.dtype,np.floating) is True:
            continue
        else:
            to_find_uniques = set(x_to_find)
            dic = {}
            j = 0
            for i in to_find_uniques:
                dic[i] = dic.get(i,0)+j
                j+=1
            for i in range(len(x_to_find)):
                if x_to_find[i] in dic:
                    x_to_find[i] = dic[x_to_find[i]]
    return data
data = find_and_convert(data)
data
```

## Try to convert all data points in X to single digits to see if it makes any progress in accuraccy of the model

```python
def convert_all(data):
    for i in data:
        x_to_convert = data[i].values
        if x_to_convert[0]/10>1:
            for i in range(len(x_to_convert)):
                x_to_convert[i] = x_to_convert[i]/10
    return data
X = convert_all(X)
```
